```python
# 파이차트

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

hat = pd.read_csv('ch4-1.csv')
pct = hat.chick / hat.chick.sum() # 부화 비율

col7 = sns.color_palette('Pastel2', 7)
plt.figure(figsize=(10, 10))
plt.pie(pct, labels=hat.hatchery, autopct='%.1f%%', colors=col7, startangle=90, counterclock=False) # 소수 첫째 자리
plt.show()
```
