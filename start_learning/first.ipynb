```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
dataset = datasets.load_iris()
X, y = dataset['data'], dataset['target'] # 여러 개의 값을 가지고 있는 경우 대문자, 하나의 값일 경우 소문자'

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0) #데이터세트 학습세트 75% 테스트 세트 25$ 비율

model = KNeighborsClassifier(n_neighbors=5) # 근처에 있는 데이터 
model.fit(X_train, y_train) # 학습

train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

# y_predicted = model.predict(X)
# print('모델이 추론한 타깃: {labels}'.format(labels=y_predicted))
# print('실제타깃 : {labels}'.format(labels=y))
print('학습 세트 정확도: {:.3f}'.format(train_score))
print('학습 세트 정확도: {:.3f}'.format(test_score))
```

```python
import matplotlib.pyplot as plt

#data = [11, 14, 12, 9, 15, 17]
x_data = [7, 8, 9, 10, 11, 12]

y1_data = [11, 14, 12, 9, 15, 17]
y2_data = [6, 8, 15, 10, 12, 11]

plt.title('Monthly sales')

plt.plot(x_data, y1_data, label='laptop')
plt.plot(x_data, y2_data, label='desktop')

plt.xlabel('Months')
plt.ylabel('Sales')

plt.legend()
plt.show()

# 꺽은선 그래프 : 데이터의 '변화를 파악하는데 유용한 그래프
# 산점도 그래프 : 두 가지 특성 사이의 '관계'를 파악하는데 유용한 그래프

```
